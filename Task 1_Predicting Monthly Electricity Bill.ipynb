```python
import numpy as np
from sklearn.linear_model import LinearRegression
```

```python
X = np.array ([[3], [4], [6], [7], [8], [10], [12]])
y = np.array ([50, 60, 80, 95, 110, 130, 150])
y
```

```python
model = LinearRegression()
model.fit (X,y)
```

```python
predicted_bill_5_hours = model.predict(np.array([[5]]))
print ("Predicted monthly bill for 5 hours of daily usage:", predicted_bill_5_hours[0])
```

```python
slope = model.coef_[0]
intercept = model.intercept_
print("Slope (increase per additional hour):", slope)
print ("Intercept (base bill):", intercept)
```

```python
import matplotlib.pyplot as plt
plt.scatter (X, y, color='blue', label = 'Data Points')
X_range = np.array([[X] for X in range (2, 14)])
y_pred = model.predict(X_range)
plt.plot(X_range, y_pred, color='red', label='Regression Line')
plt.scatter(5, predicted_bill_5_hours, color='green', 
            label='Prediction for 5 hours', s=200, edgecolor='black')
plt.title('Monthly Electricity Bill Prediction')
plt.xlabel('Average Daily Usage (hours)')
plt.ylabel('Monthly Bill (dollars)')
plt.legend ()
plt.grid(True)
plt.show()
```
