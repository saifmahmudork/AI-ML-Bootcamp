```python
import numpy as np
from sklearn.linear_model import LinearRegression
#importing necessary libraries
```

```python
X= np.array ([[850], [900], [1200], [1500], [1700], [1850], [2000], [2100]])
y= np.array ([120, 130, 150, 180, 200, 210, 230, 240])
y
#x is defining as a house size and y is defining as the price
```

```python
model = LinearRegression()
model.fit (X,y)
#initiating the train model
```

```python
LinearRegression () #Training model initiated
predictions = model.predict (X)
print (predictions)
```

```python
new_input = np.array ([[1600]]) # Create a new input (random size of a house, e.g., 1600 sq ft)
```

```python
predicted_price = model.predict (new_input)
print (f"Predicted price for a house of size {new_input [0] [0]} sq ft: {predicted_price [0] * 100:.2f}")
```

```python
new_input = np.array([[4200]])
predicted_price = model.predict(new_input)
print(f"Predicted price for a house of size {new_input[0] [0]} sq ft: ${predicted_price [0] * 1000:.2f}")
```

```python
print ("Slope (m):", model.coef_[0])
print ("Intercept (b):", model.intercept_)
```
