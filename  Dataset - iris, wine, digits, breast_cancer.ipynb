```python
import pandas as pd #importing python library like pandas to load and view data
```

```python
from sklearn.datasets import load_iris #loading a sample dataset from Scikit-Learn
data = load_iris()
df = pd.DataFrame (data.data, columns = data.feature_names)
df['target'] = data.target
df.head()
```

```python
import matplotlib.pyplot as plt #importing plots to visualize
```

 Scatter plot of sepal length and width

```python
plt.scatter (df['sepal length (cm)'], df['sepal width (cm)'], c = data.target)
```
