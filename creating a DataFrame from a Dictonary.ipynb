```python
import pandas as pd
data = {'Name': ['Alice', 'Bob', 'Catlyn'], 'Age': [24, 30, 45], 'Calls': [24, 30, 45]}
df = pd.DataFrame (data, index = ['Person1', 'Person2', 'Person3'], 
                   columns = ['Name', 'Age', 'Calls'],
                   dtype = 'str', copy = True)
df
```

```python
import matplotlib.pyplot as plt
```

```python
plt.scatter ( df['Age'], df['Calls'], c = 'blue', s = 50, alpha = .9)
```

```python
plt.scatter ( df['Age'], df['Calls'], c = 'blue', s = 50, alpha = .9)
```

```python
plt.xlabel ('Age this is')
```

```python
plt.ylabel ('This is calls')
```

```python
plt.title("It's my life")
```
